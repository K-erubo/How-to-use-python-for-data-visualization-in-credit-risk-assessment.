# Import relevant libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load data will use for credit assessment

```python
data = pd.read_csv('credit_risk_data.csv')
data
```

# Check data types for the columns

```python
print(data.dtypes)
```

# Data cleaning and conversion

```python
data['person_income'] = data['person_income'].astype(float)
data['loan_amount'] = data['loan_amount'].astype(float)
data['loan_int_rate'] = data['loan_int_rate'].astype(float)
data['debt_to_income_ratio'] = data['debt_to_income_ratio'].astype(float)
```

# Distribution of numerical values

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['person_age'], bins=40, kde=True)
plt.title('Distribution of Person Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

# Plot of Loan Amount by Loan Grade

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='loan_grade', y='loan_amount')
plt.title('Loan Amount Distribution by Loan Grade')
plt.xlabel('Loan Grade')
plt.ylabel('Loan Amount')
plt.show()
```

# Plot of Debt-to-Income Ratio vs. Interest Rate

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='debt_to_income_ratio', y='loan_int_rate', hue='cb_person_default_on_file')
plt.title('Debt-to-Income Ratio vs. Interest Rate')
plt.xlabel('Debt-to-Income Ratio')
plt.ylabel('Interest Rate')
plt.legend(title='Default')
plt.show()
```

# Plot of Home Ownership

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='person_home_ownership')
plt.title('Count of Home Ownership Types')
plt.xlabel('Home Ownership')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

# Plot person Income vs. Loan Amount


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='person_income', y='loan_amount', data=data, hue='loan_intent', palette='Dark2')
plt.xlabel('Person Income')
plt.ylabel('Loan Amount')
plt.title('Person Income vs. Loan Amount')
plt.show()
```
